# Preparation

## Import Library

```python
import pandas as pd
import numpy as np
```

## Input Dataset

```python
#Membaca data dan memasukkannya ke dalam bentuk Pandas Dataframe
df = pd.read_csv('https://raw.githubusercontent.com/Rietaros/kampus_merdeka/main/SC_HW1_bank_data.csv')
```

```python
#Jalankan code untuk mengecek nama kolom yang tersedia
df.columns
```

```python
df.head()
```

```python
#Hilangkan kolom yang dirasa tidak relevan dengan model (contoh: ID). None dapat diisi dengan nama-nama kolom yang akan digunakan.
#Contoh df = df[['X1','X2', 'Y']].copy()

#START CODE
df = df[['Geography',	'Gender',	'Age', 'Tenure',	'Balance',	'NumOfProducts',	'HasCrCard',	'IsActiveMember',	'EstimatedSalary',	'Exited']].copy()
df
```

```python
df.info()
```

## Preprocessing

```python
#Lakukan One-Hot Encoder untuk data categorical, dengan fungsi pandas get_dummies

#START CODE
df = pd.get_dummies(df, columns=["Geography", "Gender"])
df.head()
```

```python
#Pisahkan mana X (feature) dengan Y,
#Y adalah kolom "Exited"

#START CODE
X = df.drop("Exited", axis=1)
y = df["Exited"]
```

```python
X.head()
```

```python
y.head()
```

```python
#Lakukan Scaler dan/atau Noermalisasi jika diperlukan
from sklearn.preprocessing import MinMaxScaler

#START CODE
scaler = MinMaxScaler()
X_transform = scaler.fit_transform(X)
```

```python
#Ini digunakan jika dilakukan scaler/Normalisas. Jika tidak, code ini bisa dilewat dan diganti dengan code yang ada di dalam komen
X_transform = pd.DataFrame(X_transform, columns = X.columns)
#X_transform = X.copy()
```

## Train-Test Split

```python
#Split menjadi train dan test dengan test_size 25%
#Tidak perlu mengubah code ini

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_transform,y,test_size = 0.25,random_state = 123)
```

# Modeling

## Model1

### Soal :
Jelaskan secara Singkat Model pertama yang digunakan!

Model pertama yang digunakan adalah Logistic Regression. Logistic Regression adalah metode statistik yang digunakan untuk memodelkan hubungan antara variabel independen dan variabel dependen yang bersifat biner. Tujuannya adalah untuk memprediksi probabilitas kejadian suatu peristiwa berdasarkan nilai-nilai variabel independen yang diberikan. Model ini menggunakan fungsi logistik atau sigmoid untuk mengubah keluaran linier menjadi probabilitas yang berkisar antara 0 dan 1. Dengan menggunakan teknik optimasi, Logistic Regression mencari parameter yang paling sesuai dengan data pelatihan untuk menghasilkan model yang dapat digunakan untuk klasifikasi data baru.

```python
#Pilih salah satu metode Machine Leaarning
#Model Machine Learning dapat dipanggil terlebih dahulu melalui library yang digunakan. Gunakan library scikit learn seperti pada contoh

#Contoh pemanggilan library dan penggunaannya dalam model

#from sklearn.linear_model import LogisticRegression
#model1 = LogisticRegression()
#params = { "tol": [0.1,0.01,0.001], 'C':[0.5,1.0,1.5,2.0]}


#START CODE
from sklearn.linear_model import LogisticRegression
model1 = LogisticRegression()
params = { "tol": [0.1,0.01,0.001], 'C':[0.5,1.0,1.5,2.0]}
#END CODE

#Lakukan parameter tuning sesuai hyperparameter yang dibutuhkan
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(
             estimator= model1,
             param_grid= params,
             scoring = 'accuracy',
             n_jobs = 10, # core cpu yang digunakan
             cv = 10 # 3-fold cross validation (artinya kita melakukan iterasi model sebanyak 3 kali)
            )

grid.fit(X_train,y_train)
grid.best_params_
```

```python
#lakukan evaluasi dengan beberapa maetric di bawah ini
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

y_pred = grid.predict(X_test)

print(classification_report(y_test,y_pred))
print("")
print(confusion_matrix(y_test,y_pred))
print("")
print(accuracy_score(y_test, y_pred))
```

## Model2
### Soal :
Jelaskan secara Singkat Model ke-2 yang digunakan!

Model kedua yang digunakan adalah K-Nearest Neighbor (KNN) Classifier. K-Nearest Neighbor (KNN) Classifier adalah algoritma machine learning yang digunakan untuk masalah klasifikasi. Dalam KNN, prediksi dilakukan dengan mencari k tetangga terdekat dari data baru berdasarkan jarak Euclidean atau metrik lainnya. Model ini mengklasifikasikan data baru berdasarkan mayoritas label dari tetangga terdekatnya. Nilai k dapat ditentukan untuk mengatur jumlah tetangga yang digunakan dalam prediksi.  

```python
#Pilih salah satu metode Machine Leaarning
#Model Machine Learning dapat dipanggil terlebih dahulu melalui library yang digunakan. Gunakan library scikit learn seperti pada contoh

#Contoh pemanggilan library dan penggunaannya dalam model

#from sklearn.linear_model import LogisticRegression
#model1 = LogisticRegression()
#params = { "tol": [0.1,0.01,0.001], 'C':[0.5,1.0,1.5,2.0]}


#START CODE
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier(n_neighbors=1)
knn_model.fit(X_train, y_train)
#END CODE
```

```python
#lakukan evaluasi
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

y_pred = knn_model.predict(X_test)

print(classification_report(y_test,y_pred))
print("")
print(confusion_matrix(y_test,y_pred))
print("")
print(accuracy_score(y_test, y_pred))
```

## Model3
### Soal :
Jelaskan secara Singkat Model ke-3 yang digunakan!

Model ketiga yang digunakan adalah Random Forest Classifier. Random Forest Classifier adalah algoritma machine learning yang digunakan untuk masalah klasifikasi. Model ini terdiri dari beberapa pohon keputusan yang bekerja secara independen. Setiap pohon keputusan dibangun dengan menggunakan subset acak dari data pelatihan dan fitur yang dipilih secara acak. Prediksi dilakukan dengan menggabungkan hasil prediksi dari setiap pohon keputusan melalui voting mayoritas. Random Forest Classifier efektif dalam mengatasi overfitting dan dapat menghasilkan prediksi yang akurat dalam berbagai jenis data.

```python
#Pilih salah satu metode Machine Leaarning
#Model Machine Learning dapat dipanggil terlebih dahulu melalui library yang digunakan. Gunakan library scikit learn seperti pada contoh

#Contoh pemanggilan library dan penggunaannya dalam model

#from sklearn.linear_model import LogisticRegression
#model1 = LogisticRegression()
#params = { "tol": [0.1,0.01,0.001], 'C':[0.5,1.0,1.5,2.0]}


#START CODE
from sklearn.ensemble import RandomForestClassifier
model3 = RandomForestClassifier(n_estimators=390, max_features='auto', random_state=42)
#END CODE

model3.fit(X_train,y_train)
```

```python
#lakukan evaluasi
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

y_pred = model3.predict(X_test)

print(classification_report(y_test,y_pred))
print("")
print(confusion_matrix(y_test,y_pred))
print("")
print(accuracy_score(y_test, y_pred))
```
